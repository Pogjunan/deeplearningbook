# `Gradient-Based Learning`

The largest diﬀerence between the linear models we have seen so far and neuralnetworks is that the nonlinearity of a neural network causes most interesting lossfunctions to become nonconvex :
<br>

`-` linear model 과 달리 nonlinear 모델은 loss function 을 볼록하지않게 만든다.

`1`  nonlinear에서 항상 training 은 iteration 으로 이루어지며 gradient-based optimizers 는 단순히 loss function 을 낮은 cost 로 간다. (전역변수로 가지않는다)

`2` Convex 는 전역변수를 포함하는데 nonlinear 은 Convex 하지 않게 만든다.

`-` 그 중 `Stochastic gradient` 는 nonconvex 하여 수렴을 보장하지않고 초기값을 정하는 것이 매우 민감하고 중요하다.

보통 초기값으로 0~ 매우작은 양수로 가중치와 bias 를 선택한다.

지금부터 forward 와 back-propagation 이 어떻게 이루어지는 지와 cost function 을 정하는 과정을 거쳐야한다.
