# Deep Feedforward Network

`-` MLP 랑 같은 의미 multilayer perceptrons
<br>
`-` 목적 : f*(x) 로 만들기.
<br>
`-` y = f(x;$\theta$ )

Feedforward Network 로 함수를 만들어 위의 함수로 connection 을 계속 만들면 이를 `RNN` 이라한다.

# Feedforward neural networks 를 우리는 `NETWORK` 라고 부른다.

vector의역할 : 각 element 는 vector 로 구성되어있고 neuron 에 대응된다.

layer : vector to vector 함수가 아니라 뉴런의 집합으로 
많은 units 으로 구성된 vector to scalar 함수가 병렬로 구성된 형태를 띠고있다.

# 우리는 선형과 비선형함수를 쓰는데 함수를 비선형함수를 어떻게 고르지?

`-` 선형은 용량이 너무 커지는 단점이 존재한다. 비선형으로가자.

<br>
비선형함수 :  커널 기반 머신러닝 모델에서 입력 데이터를 고차원 공간으로 매핑하는 함수로 RBF kernel 에 기초한다. (RBF kernel 은 가우시안 방사 기저 함수(radial basis function)으로 불리며, 
<br>



RBF kernel 의 특성으로 무한차원으로 project 과 벡터 x 를 새로운 벡터 공간에 사영시키는 함수에 대한 벡터의 내적을 계산한다.

<br>
RBF kernel 의 무한차원으로의 증명과 전개과정 :

# 이제 그 함수를 어떻게 만들것인가?

deep learning 은 함수를 learn 하는 방법이 있다.
$ y = f(x; \theta , w ) = φ (x; \theta)^Tw. $

φ 가 히든 layer 에서 정의되는 것으로 convex할 수없는 이유이기도 하고 Feedforward 에 사용하며 이 때 `activation functions` 이 hidden layer 값 계산에 사용된다.
