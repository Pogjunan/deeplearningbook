## `Chapter 3`

`Why Probability?`
<br>
- 많은 Computer 분야에서는 결정되고 확정적이지만 `Machine Learning` 은 대부분 불확정적이고 확률적인 양을 다룹니다. <br>

## `-` 불확정에는 3가지 원인이 있다. (There are three possible sources of uncertainty)

1. 본질적 확률성 (`Inherent stochastictiy`) :  본질적으로 양자역학의 방해처럼 dynamic 하게 움직이기 떄문에 생기는 확률적인 문제. `random dynamics`으로 생겼다. (like card shuffle)

2. 완벽하지않은 관측 (`Incomplete observability`) : 모든 변수를 보지못하고 확률적으로 결정된 것만 볼 때 생기는 시스템이다. (`like 몬티홀 문제`)

3. 완벽하지 않은 모델링 (`Incomplete modeling`) : 모델을 사용할 때 관측한 정보를 버려야하고, 이 때의 버려진 정보는 모델 예측에 불확실성을 유발한다. (미래를 예측하면 불확정적으로 정확하게 한 지점을 예측할 수 없다.) (예시 : 로봇에게 환경에 따라 어떠한 지점으로 움직이는 학습을 시키고 `prediction` 을 진행하면 한 지점에 대한 확정적 표시가 아닌 불확실한 위치정보를 예측하게되어 확률적이라 할 수 있다.)

`Random Variables(확률변수)`  <br>
`-` 어떤 사건에 대해 숫자로서 정의내릴 수 있는 변수 
<br>(예시: 동전 던져서 앞면이 나오는 개수를 확률변수 $X$ 라 한다면 $X=0$ `: 앞면이 0번 나온 사건 , `$P(X=0)$` : 앞면이 0번 나온 확률 등 ...`

`Probability distribution (확률분포)` <br>
`-`  확률 변수가 특정한 값을 가질 확률을 나타내는 함수 (`f(x) = 수식` 의 notation 을 이용하고 이 때 f(x) 는 확률밀도함수(pdf)이다.) 
<br>
`-` 확률질량함수인 경우 f(x) 에 대한 정의는 이산확률변수 X 에 대해 `f(x) = P(X=x)` 이다. 이산확률변수 X 에 대한 함수 = 확률질량함수
<br>
`-` 확률밀도함수인 경우 f(x) 에 대한 정의는 연속확률변수 X 에 대해  이다.
<br>
`-` 연속확률변수 X 에 대한 함수 = 확률밀도함수 
<br>
또는 각자의 고유 함수 표현식을 사용한다. (균등인 경우 : $U(a,b)$ , 정규분포인 경우 : $N(μ,σ)$
