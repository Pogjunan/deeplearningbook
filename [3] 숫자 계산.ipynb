## `-` Numerical Computation (수치계산)

`OverFlow and Underflow`

Underflow : 보통 숫자가 `0` 근처에서 반올림할 때 생긴다. 또 `0` 근처에서 `log` 를 취하면 -무한대로 계산되어 문제가 발생한다.

Overflow : 숫자가 무한 , -무한 과 같은 큰 크기의 숫자를 만났을 때 생긴다.

`SoftMax function` : Underflow 와 Overflow 가 안정화되어야하는 함수

$x_i = c$ 로 잡았을 때 c 의 값에 따라 너무 크면 $exp(c)$ 는 Overflow 가 되고 , c 가 작고 음수이면 $exp(c)$ 가 Underflow 를 만들 것입니다.

`softmax 함수 안정화하는 방법` : 입력 벡터 x에 대해 각 요소에서 최대값을 빼어 새로운 벡터 z를 생성한 후 softmax 함수를 이 새로운 벡터 z에 적용시에  exp 함수의 가장 큰 숫자가 0이 되므로 오버플로우(Overflow)가 발생하지않습니다. 여전히 분자의 Underflow 가 발생할 수 있습니다.  log softmax(x) 를 이용할 수 있고 단순히 변환시에는 로그 함수에서 -∞ 가 나올 수 있으므로 별도의 함수를 만들어서 해결이 가능합니다.
