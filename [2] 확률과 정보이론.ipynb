## `-` 불확정에는 3가지 원인이 있다. (There are three possible sources of uncertainty)

1. 본질적 확률성 (`Inherent stochastictiy`) :  본질적으로 양자역학의 방해처럼 dynamic 하게 움직이기 떄문에 생기는 확률적인 문제. `random dynamics`으로 생겼다. (like card shuffle)

2. 완벽하지않은 관측 (`Incomplete observability`) : 모든 변수를 보지못하고 확률적으로 결정된 것만 볼 때 생기는 시스템이다. (`like 몬티홀 문제`)
